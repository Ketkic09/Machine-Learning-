To predict per capita income of Canada using Linear Regression

```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```
df = pd.read_csv('/content/sample_data/canada_per_capita_income.csv')
df
```

```
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income(US $)')
plt.scatter(df.year,df.income, color='green', marker='*')

```

```
new_df= df.drop('income',axis='columns')
new_df
```

```
income= df.income
income
```

Creating Regression Model and fitting our data using fit()

```
reg=linear_model.LinearRegression()
reg.fit(new_df,income)
```

Plotting our best fit line

```
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income(US $)')
plt.scatter(df.year,df.income, color='green', marker='*')
plt.plot(df.year,reg.predict(df[['year']]), color='blue')
```

Lets predict value for the year 2020

```
reg.predict([[2020]])
```

```
test_df= pd.read_csv('/content/sample_data/income_test.csv')
test_df
```

```
pred= reg.predict(test_df)
pred
```

```
test_df['income'] = pred
test_df
```

Plotting best fit line for our test data set

```
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income(US $)')
plt.scatter(test_df.year,test_df.income, color='green', marker='*')
plt.plot(test_df.year,pred, color='blue')
```

```
m=reg.coef_
c=reg.intercept_
xs= test_df.year
```

```
#Caluculating regression line using the line eqaution 
regression_line = [(m*x)+c for x in xs]
regression_line
```
